# Pretrained ResNet50 for Image Classification

```python
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = ResNet50(weights='imagenet')
```

```python
img_path = 'Golde33443.jpg'
```

```python
img = image.load_img(img_path,target_size=(224,224))
```

```python
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array,axis=0)
img_array = preprocess_input(img_array)
```

```python
predictions = model.predict(img_array)
```

```python
decoded_predictions = decode_predictions(predictions,top=5)[0]
for i , (imagenet_id, label, score) in enumerate(decoded_predictions):
    print(f"{i+1}: {label} ({score:.2f})")
```

# Fine-Tuning ResNet50 on custom Dataset

```python
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from tensorflow.data import Dataset

# Load CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

# Normalize pixel values (0 to 1)
x_train, x_test = x_train / 255.0, x_test / 255.0

# Convert labels to one-hot encoding
y_train, y_test = to_categorical(y_train, 10), to_categorical(y_test, 10)

# Function to resize images in dataset
def preprocess(image, label):
    image = tf.image.resize(image, (224, 224))  # Resize to ResNet50 input size
    return image, label

# Convert datasets into TensorFlow Dataset objects (efficient batch processing)
BATCH_SIZE = 32
train_dataset = Dataset.from_tensor_slices((x_train, y_train)).map(preprocess).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
test_dataset = Dataset.from_tensor_slices((x_test, y_test)).map(preprocess).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

# Load ResNet50 without top layers
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze pretrained layers
for layer in base_model.layers:
    layer.trainable = False

# Add custom layers
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(256, activation='relu')(x)
x = Dense(128, activation='relu')(x)
output_layer = Dense(10, activation='softmax')(x)  # CIFAR-10 has 10 classes

# Create the model
model = Model(inputs=base_model.input, outputs=output_layer)

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])

# Train model
model.fit(train_dataset, validation_data=test_dataset, epochs=10)

# Fine-tuning: Unfreeze last 10 layers and re-train
for layer in base_model.layers[-10:]:
    layer.trainable = True

# Compile again with a lower learning rate
model.compile(optimizer=Adam(learning_rate=0.00001), loss='categorical_crossentropy', metrics=['accuracy'])

# Continue training
model.fit(train_dataset, validation_data=test_dataset, epochs=10)

# Evaluate model
test_loss, test_acc = model.evaluate(test_dataset)
print(f"Test Accuracy: {test_acc:.4f}")
```
