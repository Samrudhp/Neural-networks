```python
import tensorflow as tf
from tensorflow.keras.layers import Normalization
```

```python
normalizer = Normalization(axis=-1)
```

```python
import numpy as np
data = np.array([[10.0,20.0],[30.0,40.0],[50.0,60.0]])

normalizer.adapt(data)# calculate mean and variance4
```

```python
normalized_data = normalizer(data) # apply normalization
print(normalized_data.numpy())
```

```python
from  tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Normalization(axis = -1),
    Dense(64,activation="relu"),
    Dense(1) # regression output
])

model.layers[0].adapt(data)

model.compile(optimizer = "adam",loss="mse")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=1000)  # Mean=50, Std Dev=15
```

```python
mean = np.mean(data)
std = np.std(data)
normalized_data = (data - mean) / std  # Standardization
```

```python
# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Original Data Distribution
sns.histplot(data, bins=30, kde=True, color='blue', ax=axes[0])
axes[0].axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')
axes[0].set_title("Original Data Distribution")
axes[0].set_xlabel("Value")
axes[0].set_ylabel("Frequency")
axes[0].legend()

#normalized data distribution

sns.histplot(normalized_data, bins=30, kde=True, color='green', ax=axes[1])
axes[1].axvline(0, color='red', linestyle='dashed', linewidth=2, label="Mean: 0")
axes[1].set_title("Normalized Data (Zero Mean, Unit Variance)")
axes[1].set_xlabel("Value")
axes[1].set_ylabel("Frequency")
axes[1].legend()

plt.tight_layout()
plt.show()



```
